```python
import seaborn as sns
import plotly.graph_objects as go
titanic=sns.load_dataset("titanic")
fig=go.Figure()
plot=fig.add_trace(go.Scatter(x=titanic.age,y=titanic.fare,mode="markers"))
```

```python
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a box plot
fig = px.box(tips, x="day", y="total_bill", color="sex")

```

```python
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a histogram
fig = px.histogram(tips, x="day", y="total_bill", color="sex",
                   pattern_shape="smoker")
```

```python
import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Create the scatter matrix plot
fig = px.scatter_matrix(
    iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species"
)
```

```python
'''  distplot is a function used to visualize the distribution of a univariate dataset in Python. 
It plots a histogram of the data along with a continuous probability density function (PDF) estimate. 
It is a popular visualization technique in data analysis and statistics.'''

import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Create a distplot of the "total_bill" column
fig = px.histogram(tips, x="total_bill",opacity=0.7, color='sex')
```
